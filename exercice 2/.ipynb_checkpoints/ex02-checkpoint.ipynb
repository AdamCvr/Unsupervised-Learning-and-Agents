# Part 2 : meteorological data : dimensionality reduction and visualization


# Perform Dimensionality Reduction and Visualization


Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

Load Data and Labels

```python
data = np.load('data.npy')
labels = np.load('labels.npy')
```

Before applying a PCA, it is necessary to be sure that the data here is standardized

```python
data_std = StandardScaler().fit_transform(data)
```

Perform PCA (Principal Component Analysis) for dimensionality reduction to 2D then in 3D. The main goal of PCA here is to transform high-dimensional data into a lower-dimensional

```python
# 2D
pca_2d = PCA(n_components=2)
data_pca_2d = pca_2d.fit_transform(data_std)

# 3D
pca_3d = PCA(n_components=3)
data_pca_3d = pca_3d.fit_transform(data_std)
```

We then perform t-SNE (t-Distributed Stochastic Neighbor Embedding) for dimensionality reduction to 2D then in 3D. this technique is used for visualizing high-dimensional data in a lower-dimensional space. 

```python
# Perform t-SNE for dimensionality reduction to 2D
tsne_2d = TSNE(n_components=2)
data_tsne_2d = tsne_2d.fit_transform(data_std)

# Perform t-SNE for dimensionality reduction to 3D
tsne_3d = TSNE(n_components=3)
data_tsne_3d = tsne_3d.fit_transform(data_std)
```

We then plot the results in 2D projections then in 3D

```python

# Plot 2D projections
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data_pca_2d[:, 0], data_pca_2d[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('PCA 2D Projection')

plt.subplot(1, 2, 2)
plt.scatter(data_tsne_2d[:, 0], data_tsne_2d[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('t-SNE 2D Projection')

plt.show()

# Plot 3D projections
fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.scatter(data_pca_3d[:, 0], data_pca_3d[:, 1], data_pca_3d[:, 2], c=labels, cmap='viridis', edgecolors='k')
ax1.set_title('PCA 3D Projection')

ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax2.scatter(data_tsne_3d[:, 0], data_tsne_3d[:, 1], data_tsne_3d[:, 2], c=labels, cmap='viridis', edgecolors='k')
ax2.set_title('t-SNE 3D Projection')

plt.show()
```
