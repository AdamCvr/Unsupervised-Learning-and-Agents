We load Data and Labels

```python
data = np.load('data.npy')
labels = np.load('labels.npy')
```


Before we continue with the UMAP let's standardize the Data
UMAP is used for its ability to capture non-linear relationships


```python
data_std = StandardScaler().fit_transform(data)
```

We perform UMAP for dimensionality reduction to 2D and 3D


```python
umap_2d = UMAP(n_components=2)
data_umap_2d = umap_2d.fit_transform(data_std)

umap_3d = UMAP(n_components=3)
data_umap_3d = umap_3d.fit_transform(data_std)
```

Here we plot as 2D projections and 3D projections



```python
#2D projections
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data_umap_2d[:, 0], data_umap_2d[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('UMAP 2D Projection')

#3D projections
fig = plt.figure(figsize=(18, 6))

ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax1.scatter(data_umap_3d[:, 0], data_umap_3d[:, 1], data_umap_3d[:, 2], c=labels, cmap='viridis', edgecolors='k')
ax1.set_title('UMAP 3D Projection')

plt.show()
```
